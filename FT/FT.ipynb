```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(-6, 6, .01)
hz = np.random.normal(0, 3, x.shape)
```

```python
# y = np.where((x < np.pi) & (x > -np.pi), 10*np.sin(x)+np.sin(8*x)+np.sin(50*x), 0)
# y = np.where((x < np.pi) & (x > -np.pi), 10*np.sin(x)+np.sin(8*x), 0)
y = np.where((x < np.pi) & (x > -np.pi), 10*np.sin(x), 0)+hz
# y = 10*np.sin(x)+np.sin(8*x)+np.sin(50*x)
# y = np.where((x < np.pi) & (x > -np.pi), 10*np.sin(50*x), 0)
# y = np.zeros(x.shape)
```

```python
# z = np.where((x < 3/2*np.pi) & (x > -np.pi/2), 10*np.sin(x-np.pi/2)+np.sin(8*(x-np.pi/2))+np.sin(50*(x-np.pi/2)), 0)
# z = np.where((x < 3/2*np.pi) & (x > -np.pi/2), 10*np.sin(x-np.pi/2)+np.sin(8*(x-np.pi/2)), 0)
z = np.where((x < 3/2*np.pi) & (x > -np.pi/2), 10*np.sin(x-np.pi/2), 0)+hz
# z = np.where((x < 3/2*np.pi) & (x > -np.pi/2), 10*np.sin(x-np.pi/2), 0)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 9))
axs[0].plot(x, y)
axs[1].plot(x, z)
```

```python
Ty = np.fft.fft(y)
Ty = Ty/y.size
Tz = np.fft.fft(z)
Tz = Tz/y.size
```

```python
fig = plt.figure(figsize=(20, 10))
ax11 = fig.add_subplot(3, 4, 1)
ax12 = fig.add_subplot(3, 4, 2, projection='3d')
ax13 = fig.add_subplot(3, 4, 3)
ax14 = fig.add_subplot(3, 4, 4)

ax21 = fig.add_subplot(3, 4, 5)
ax22 = fig.add_subplot(3, 4, 6, projection='3d')
ax23 = fig.add_subplot(3, 4, 7)
ax24 = fig.add_subplot(3, 4, 8)

ax31 = fig.add_subplot(3, 4, 9)
ax32 = fig.add_subplot(3, 4, 10, projection='3d')
ax33 = fig.add_subplot(3, 4, 11)
ax34 = fig.add_subplot(3, 4, 12)

mxy = np.max(np.abs(Ty))
ax12.set_ylim([-mxy, mxy])
ax12.set_zlim([-mxy, mxy])
ax13.set_ylim([-mxy, mxy])
ax14.set_ylim([-mxy, mxy])
ax11.plot(np.abs(Ty))
ax12.plot(np.arange(Ty.size), np.real(Ty), np.imag(Ty))
ax13.plot(np.real(Ty))
ax14.plot(np.imag(Ty))

mxz = np.max(np.abs(Tz))
ax22.set_ylim([-mxz, mxz])
ax22.set_zlim([-mxz, mxz])
ax23.set_ylim([-mxz, mxz])
ax24.set_ylim([-mxz, mxz])
ax21.plot(np.abs(Tz))
ax22.plot(np.arange(Tz.size), np.real(Tz), np.imag(Tz))
ax23.plot(np.real(Tz))
ax24.plot(np.imag(Tz))

mxyz = np.max([mxy, mxz])
ax31.set_ylim([-.05, mxyz])
ax32.set_ylim([-mxyz, mxyz])
ax32.set_zlim([-mxyz, mxyz])
ax33.set_ylim([-mxyz, mxyz])
ax34.set_ylim([-mxyz, mxyz])
ax31.plot(np.abs(Ty)-np.abs(Tz))
ax32.plot(np.arange(Tz.size), np.real(Ty-Tz), np.imag(Ty-Tz))
ax33.plot(np.real(Ty-Tz))
ax34.plot(np.imag(Ty-Tz))
```

```python
Ty = np.where(np.abs(Ty) > 1, Ty, 0)
Tz = np.where(np.abs(Tz) > 1, Tz, 0)

fig = plt.figure(figsize=(20, 10))
ax11 = fig.add_subplot(3, 4, 1)
ax12 = fig.add_subplot(3, 4, 2, projection='3d')
ax13 = fig.add_subplot(3, 4, 3)
ax14 = fig.add_subplot(3, 4, 4)

ax21 = fig.add_subplot(3, 4, 5)
ax22 = fig.add_subplot(3, 4, 6, projection='3d')
ax23 = fig.add_subplot(3, 4, 7)
ax24 = fig.add_subplot(3, 4, 8)

ax31 = fig.add_subplot(3, 4, 9)
ax32 = fig.add_subplot(3, 4, 10, projection='3d')
ax33 = fig.add_subplot(3, 4, 11)
ax34 = fig.add_subplot(3, 4, 12)

mxy = np.max(np.abs(Ty))
ax12.set_ylim([-mxy, mxy])
ax12.set_zlim([-mxy, mxy])
ax13.set_ylim([-mxy, mxy])
ax14.set_ylim([-mxy, mxy])
ax11.plot(np.abs(Ty))
ax12.plot(np.arange(Ty.size), np.real(Ty), np.imag(Ty))
ax13.plot(np.real(Ty))
ax14.plot(np.imag(Ty))

mxz = np.max(np.abs(Tz))
ax22.set_ylim([-mxz, mxz])
ax22.set_zlim([-mxz, mxz])
ax23.set_ylim([-mxz, mxz])
ax24.set_ylim([-mxz, mxz])
ax21.plot(np.abs(Tz))
ax22.plot(np.arange(Tz.size), np.real(Tz), np.imag(Tz))
ax23.plot(np.real(Tz))
ax24.plot(np.imag(Tz))

mxyz = np.max([mxy, mxz])
ax31.set_ylim([-.05, mxyz])
ax32.set_ylim([-mxyz, mxyz])
ax32.set_zlim([-mxyz, mxyz])
ax33.set_ylim([-mxyz, mxyz])
ax34.set_ylim([-mxyz, mxyz])
ax31.plot(np.abs(Ty)-np.abs(Tz))
ax32.plot(np.arange(Tz.size), np.real(Ty-Tz), np.imag(Ty-Tz))
ax33.plot(np.real(Ty-Tz))
ax34.plot(np.imag(Ty-Tz))
```
