```python
import numpy as np
from PIL import Image
import cv2
from matplotlib import pyplot as plt
```

```python
img = Image.open('height_shadow00.png')
A = np.asarray(img)
```

```python
plt.imshow(A)
plt.axis('off')
```

```python
gkernel = cv2.getGaussianKernel(31, 0)
```

```python
gk2 = np.outer(gkernel, gkernel)
```

```python
B = cv2.filter2D(A, -1, gk2)
plt.imshow(B)
```

```python
plt.imshow(A.astype('int16') - B.astype('int16') + 128)
```

```python
C = A[:,:,0].copy()
D = C.copy()
plt.imshow(D)
```

```python
intv = 5
for i in range(C.shape[0]):
    for j in range(C.shape[1]):
        left = max(j - intv, 0)
        right = min(j + intv + 1, C.shape[1])
        up = max(i - intv, 0)
        down = min(i + intv + 1, C.shape[0])
        if np.all(C[up:down, left:right] <= 42):
            D[i,j] = 0
plt.imshow(D)
```

```python
E = cv2.filter2D(D, -1, gk2)
plt.imshow(E)
```

```python
new_img = Image.fromarray(np.array([E, E, E, 255*np.ones(E.shape).astype('uint8')]).transpose([1, 2, 0]))
```

```python
new_img.save('height_shadow_deep_blur.png')
```
