```python
import os
import numpy as np
from matplotlib import pyplot as plt
```

```python
nsoom = 40 # 숨 횟수
```

```python
# 로아 내 기본변수
min_s_rate = 1 # 최소확률
max_s_rate = 2 # 최대확률

gahomax = 8 # 가호 최대 개수
```

```python
# 로아 내 강화재료 가격
stone_gold = .7 # 파괴수호석 (1개당)
debri_gold = 368/1500 # 파편 (1개당)
dolpa_gold = 30 # 돌파석 (1개당)
oreha_gold = 12 # 오레하 (1개당)

# 로아 내 추가재료 가격 (추후 반영)
gahogold = 166
chukgold = 77
eungold = 31
```

```python
# 재련당 재료 투하
stones = 1300
debris = 1432
dolpa = 34
oreha = 22
gold = 750 # 재련 당 골드
```

```python
# 로아 내 기본상수
k_rate = .465
step_s_rate = (max_s_rate - min_s_rate) / 10

chukmax = gahomax * 3
eunmax = gahomax * 6

gaho_rate, chuk_rate, eun_rate = min_s_rate / 3 / gahomax, min_s_rate / 3 / chukmax, min_s_rate / 3 / eunmax

spending = stone_gold*stones + debri_gold*debris + dolpa_gold*dolpa + oreha_gold*oreha + gold # 강화당 소모 골드
spending_full = spending + gahogold*gahomax + chukgold*chukmax + eungold*eunmax # 강화당 소모 골드(풀숨)
```

```python
s_rate = np.concatenate((np.arange(min_s_rate, max_s_rate, step_s_rate), np.ones(240) * max_s_rate)) / 100 # 각 시도별 성공확률

s_rates = np.ones((s_rate.shape[0]-nsoom+1, 1)) @ s_rate.reshape((1, -1))
```

```python
for i in range(s_rates.shape[0]):
    s_rates[i] += np.concatenate((np.zeros(i), np.ones(nsoom), np.zeros(s_rates.shape[1]-nsoom-i)))*(max_s_rate - min_s_rate) / 100
```

```python
k = k_rate * np.concatenate(([0], s_rate[:-1])) # 각 시도별 장인의 기운 누적량

ks = k_rate * np.concatenate((np.zeros(s_rates.shape[0]).reshape((-1, 1)), s_rates[:,:-1]), axis = 1) # 각 시도별 장인의 기운 누적량
```

```python
s_rate_mat = np.diag(s_rate)

for i in range(s_rate_mat.shape[0]):
    for j in range(i):
        s_rate_mat[i,j] = s_rate_mat[j,j]

all_success = 1 - (1 - s_rate_mat).prod(axis = 1) # i+1 회 이내에 강화성공확률
```

```python
s_rates_mat = np.zeros((*s_rates.shape, s_rates.shape[1]))
all_successes = np.zeros(s_rates.shape)
for i in range(s_rates_mat.shape[0]):
    s_rates_mat[i] = np.diag(s_rates[i])

    for ii in range(s_rates_mat[i].shape[0]):
        for iii in range(ii):
            s_rates_mat[i,ii,iii] = s_rates_mat[i,iii,iii]

    all_successes[i] = 1 - (1 - s_rates_mat[i]).prod(axis = 1) # i+1 회 이내에 강화성공확률(풀숨)
```

```python
k_rate_mat = np.diag(k)
for i in range(k_rate_mat.shape[0]):
    for j in range(i):
        k_rate_mat[i,j] = k[j]
nth_kiwoon = k_rate_mat.sum(axis = 1) # i+1 회시에 장인의 기운 누적량
```

```python
k_rates_mat = np.zeros((*s_rates.shape, s_rates.shape[1]))
nth_kiwoons = np.zeros(s_rates.shape)

for i in range(k_rates_mat.shape[0]):
    k_rates_mat[i] = np.diag(ks[i])
    for ii in range(k_rates_mat[i].shape[0]):
        for iii in range(ii):
            k_rates_mat[i,ii,iii] = ks[i,iii]
    nth_kiwoons[i] = k_rates_mat[i].sum(axis = 1) # i+1 회시에 장인의 기운 누적량(풀숨)
```

```python
ki100 = (nth_kiwoon < 1).sum() + 1 # 장기백 달성 강화 시도 횟수
ki100s = (nth_kiwoons < 1).sum(axis = 1) + 1 # 장기백 달성 강화 시도 횟수(풀숨)
```

```python
# 장기백 사이즈로 결과값 재조정
all_success = all_success[:ki100]
nth_kiwoon = nth_kiwoon[:ki100]
s_rate_mat = s_rate_mat[:ki100, :ki100]

all_success[-1] = 1 # 마지막 재련은 강화확률 1
```

```python
# 장기백 사이즈로 결과값 재조정(풀숨)
t = ki100s.max()
all_successes = all_successes[:, :t]
nth_kiwoons = nth_kiwoons[:, :t]
s_rates_mat = s_rates_mat[:, :t, :t]

for i in range(all_successes.shape[0]):
    all_successes[i, ki100s[i]-1] = 1 # 마지막 재련은 강화확률 1
    all_successes[i, ki100s[i]:] = np.nan # 마지막 재련은 강화확률 1
    nth_kiwoons[i, ki100s[i]:] = np.nan
```

```python
meanmat = 1 - s_rate_mat
meanmat[np.diag_indices(meanmat.shape[0])] = 1 - meanmat[np.diag_indices(meanmat.shape[0])]
meanmat[-1, -1] = 1

nth_success = meanmat.prod(1) # 정확히 i+1 회에 강화성공할 확률
mean_times = (nth_success * np.arange(1, 1+len(nth_success))) # 각 강화횟수별 기대값
```

```python
meanmats = 1 - s_rates_mat
nth_successes = np.zeros(meanmats.shape[:2])
mean_timeses = np.zeros(meanmats.shape[:2])
for i in range(meanmats.shape[0]):
    meanmats[i][np.diag_indices(meanmats.shape[1])] = 1 - meanmats[i][np.diag_indices(meanmats.shape[1])]
    meanmats[i, ki100s[i]-1, ki100s[i]-1] = 1

    nth_successes[i] = meanmats[i, :, :ki100s[i]].prod(1) # 정확히 i+1 회에 강화성공할 확률(풀숨)
    nth_successes[i, ki100s[i]:] = np.nan
    mean_timeses[i] = (nth_successes[i] * np.arange(1, 1+len(nth_successes[i]))) # 각 강화횟수별 기대값(풀숨)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 6))

ax[0].plot(nth_success, '.-')
ax[0].set_title('The success rate of n-th try')
ax[0].set_xlabel('Achieved Rate of \'ki100\' : {:.5f}%\n\'ki100\' : {}'.format(nth_success[-1]*100, ki100))

for i in range(nth_successes.shape[0]):
    ax[1].plot(nth_successes[i], '.-')
    print('Achieved Rate of \'ki100\' : {:.5f}%,  \'ki100\' : {}'.format(nth_successes[i,ki100s[i]-1]*100, ki100s[i]))
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 6))

ax[0].plot(np.arange(1, len(mean_times)+1), mean_times, '.-')
ax[0].set_title('The Expactation of n-th try')
ax[0].set_xlabel('Mean Smithing Times : {}'.format(mean_times.sum())) # 평균 기대 횟수

for i in range(mean_timeses.shape[0]):
    ax[1].plot(np.arange(1, len(mean_timeses[i])+1), mean_timeses[i], '.-')
    print('Mean Smithing Times : {}'.format(mean_timeses[i][:ki100s[i]].sum())) # 평균 기대 횟수
```

```python
fig, ax = plt.subplots(1, 2, figsize = (14, 6))

ax[0].plot(all_success, '.-')
ax[0].set_title('Cumulative Success')

for i in range(all_successes.shape[0]):
    ax[1].plot(all_successes[i], '.-')
ax[1].set_title('Cumulative Success')
```

```python
print('골드 소모 기대값 : {}'.format(mean_times.sum()*spending))
print('장기백시 골드 소모 : {}'.format(ki100*spending))
for i in range(mean_timeses.shape[0]):
    print()
    print('골드 소모 기대값(전략번호 : {}) : {}'.format(i, mean_timeses[0][~np.isnan(mean_timeses[0])].sum()*spending + nsoom*(spending_full - spending)))
    print('장기백시 골드 소모(전략번호 : {}) : {}'.format(i, ki100s[i]*spending + nsoom*(spending_full - spending)))
```
