```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-20, 20, 10000)

y = np.sin(x)

z = np.fft.fft(y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(np.abs(z))
```

```python
x = np.linspace(-20, 20, 10000)

y = np.array([2*(i%2) - 1 for i in range(10000)])

z = np.fft.fft(y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(np.abs(z))
```
