```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import json
from pprint import pprint
```

```python
img = cv2.imread("123.jpg", cv2.IMREAD_UNCHANGED)
print(img.shape)
img0 = img[:,:,0]
```

```python
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(111)
ax1.imshow(cv2.cvtColor(img0, cv2.COLOR_BGR2RGB))
```

```python
A = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
T = np.zeros((img0.shape[0]-2,img0.shape[1]-2))
for i in range(T.shape[0]):
    for j in range(T.shape[1]):
        T[i,j] = np.sum(A*img0[i:i+3,j:j+3])
```

```python
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(111)
ax1.imshow(np.abs(T), cmap='Greys')
```

```python
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(111)
ax1.imshow(np.abs(T), cmap='Greys')
```

```python
A = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
T = np.zeros((img0.shape[0]-2,img0.shape[1]-2,3))
for k in range(3):
    for i in range(T.shape[0]):
        for j in range(T.shape[1]):
            T[i,j,k] = np.sum(A*img[i:i+3,j:j+3,k])
```

```python
T = np.abs(T*255/np.max(T))
T = T.astype(np.uint8)
fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(111)
ax1.imshow(cv2.cvtColor(T, cv2.COLOR_BGR2RGB))
```
