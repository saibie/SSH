```python
%matplotlib notebook

import numpy as np
from numpy import linalg as nla
import scipy as sp
from scipy import linalg as sla
from matplotlib import pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
from pprint import pprint
from time import time
import colorsys as csys
# import SHFunctions as shf
import os

from jupyterthemes import jtplot
jtplot.style(theme='solarizedl')

from IPython.core.display import display, HTML
np.set_printoptions(linewidth = 270)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1, projection='3d')
H = np.linspace(0,1,36)
S = np.linspace(0,1,10)
V = np.linspace(0,1,10)

x, y, z, c = [], [], [], []
for h in H:
    for s in S:
        for v in V:
            rgb = list(csys.hsv_to_rgb(h,s,v))
            x.append(s*np.cos(2*np.pi*h))
            y.append(s*np.sin(2*np.pi*h))
            z.append(v)
            c.append(rgb)
#             ax.scatter(rgb[0], rgb[1], rgb[2], c = [rgb])

def init():
    ax.scatter(x, y, z, c = c)
    return fig,

def animate(i):
    ax.view_init(elev=30., azim=i)
    return fig,

# Animate
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=np.arange(-180, 180, 1), interval=100, repeat_delay = 500, blit=True)
```
