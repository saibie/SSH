```python
%matplotlib notebook

import numpy as np
from numpy import linalg as nla
import scipy as sp
from scipy import linalg as sla
from matplotlib import pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
from pprint import pprint
from time import time
# import SHFunctions as shf
import os

from jupyterthemes import jtplot
jtplot.style(theme='solarizedl')

from IPython.core.display import display, HTML
np.set_printoptions(linewidth = 270)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
m = 6
e = np.linspace(0, 5, 120)
me = np.max(e)

def EigPoint(e, m):
#     A = np.ones((m, m)) - np.eye(m)
    A = np.zeros((m,m))
    for i in range(-1, m-1):
        A[i, i+1] = e
        A[i+1, i] = e
    d, X = nla.eig(A)
    return np.real(d), np.imag(d), A


ims = []
for i, index in zip(e, range(len(e))):
    im, = ax.plot(EigPoint(i, m)[0], EigPoint(i, m)[1], 'ro')
    if index == len(e) - 1:
        xmin, xmax = ax.get_xlim()
        ymin, ymax = ax.get_ylim()
        kkk = max(abs(xmin), abs(ymin), abs(xmax), abs(ymax))
        ax.set_xlim(-kkk, kkk)
        ax.set_ylim(-kkk, kkk)
    ims.append([im])
    
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=False, repeat_delay=100, repeat = True)
# plt.show()
# print(ims)
```
