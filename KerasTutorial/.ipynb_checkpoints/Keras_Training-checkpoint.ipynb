```python
import numpy as np
np.random.seed(123)
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Input
from keras.layers import Convolution2D, MaxPooling2D
from keras.models import Model
from keras.utils import np_utils
from keras.datasets import mnist
from keras.applications import mobilenet
import tensorflow as tf
from matplotlib import pyplot as plt
```

```python
A = np.random.random((1000, 200, 200, 3))
b = np.random.random((1000, 1024))
for i in range(b.shape[0]):
    b[i,:] = b[i,:] / np.sum(b[i,:])
```

```python
I = Input(shape = (200, 200, 3))
```

```python
x = mobilenet.MobileNet(alpha = 1, weights = None, include_top = False, input_tensor = I, pooling='avg',)
```

```python
model = Model(I, x.output)
model.summary()
```

```python
optimizer=keras.optimizers.Adam()
model.compile(loss=keras.losses.categorical_crossentropy, optimizer=optimizer)
```

```python
model.fit(A, b, epochs = 200)
```

```python
k = model.predict(np.random.random((1, 200, 200, 3)))
```

```python
print(k)
print(np.argmax(k))
print(np.max(k))
print(k[0, np.argmax(k)])
```
